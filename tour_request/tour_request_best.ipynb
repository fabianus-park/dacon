```python
import pandas as pd
# train = pd.read_csv('data/train.csv')
train = pd.read_csv('train_preprocessed.csv')
```

```python
from pycaret.datasets import get_data 
from pycaret.classification import *
from pycaret.utils import check_metric
```

```python
# setup으로 init
clf = setup(data=train, target='ProdTaken', use_gpu=True)
```

```python
# ML 모델 성능 비교
best = compare_models()
```

```python
print(best)
```

```python
# analyze model
evaluate_model(best)
```

```python
# test 데이터 불러오기
# test = pd.read_csv('data/test.csv')
test = pd.read_csv('test_preprocessed.csv')
```

```python
prediction = predict_model(best)
```

```python
check_metric(prediction['ProdTaken'], prediction['Label'], metric = 'Accuracy')
```

```python
# 테스트셋 predict
prediction = predict_model(best, data=test)
```
